# Calculating eigenvectors and eigenvalues for problem 1 of homework 3

```python
from sympy import *
init_printing(use_unicode=True)
```

## Part A

### First case

```python
x = symbols('delta')
```

```python
M = Matrix([[0, 1,], [-1,  -x,]])
M
```

```python
M.eigenvals()
```

```python
print(latex(M.eigenvals()))
```

### Second case

```python
M = Matrix([[0, 1,], [1,  -x,]])
M
```

```python
M.eigenvals()
```

```python
print(latex(M.eigenvals()))
```

### Second case

```python
x, y = symbols('beta delta')
```

```python
M = Matrix([[0, 1,], [-sqrt(1-x**2),  -y,]])
M
```

```python
M.eigenvects()
```

```python
print(latex(M.eigenvects()))
```
