# Calculating eigenvectors and eigenvalues for problem 1 of homework 2

```python
from sympy import *
init_printing(use_unicode=True)
```

## Part A

```python
x = symbols('epsilon')
```

```python
M = Matrix([[0, 1,], [-2,  -x,]])
M
```

```python
M.eigenvects()
```

```python
print(latex(M.eigenvects()))
```

```python
expand(Transpose(M.eigenvects()[0][-1][-1])*M.eigenvects()[1][-1][-1])
```

```python
M = Matrix([[0, 1,], [1,  -x,]])
M
```

```python
M.eigenvects()
```

```python
print(latex(M.eigenvects()))
```

```python
expand(Transpose(M.eigenvects()[0][-1][-1])*M.eigenvects()[1][-1][-1])
```
