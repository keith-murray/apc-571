# Stream plots for systems in problem 1

```python
import matplotlib.pyplot as plt
import numpy as np
```

## Part A

```python
fig, axs = plt.subplots(2, 2, figsize=(8, 6),)
axs = axs.flat

## Positive epsilon
eps = 0.5
w = 2
Y, X = np.mgrid[-w:w:100j, -w:w:100j]
U = Y
V = X - X**3 - eps*Y
speed = np.sqrt(U**2 + V**2)

strm = axs[0].streamplot(X, Y, U, V, density=[0.5, 1], color=U, cmap='autumn')
axs[0].set_title(f'Epsilon of {eps}')

## Zero epsilon
eps = 0.0
w = 2
Y, X = np.mgrid[-w:w:100j, -w:w:100j]
U = Y
V = X - X**3 - eps*Y
speed = np.sqrt(U**2 + V**2)

strm = axs[1].streamplot(X, Y, U, V, density=[0.5, 1], color=U, cmap='autumn')
fig.colorbar(strm.lines)
axs[1].set_title(f'Epsilon of {eps}')

## Negative epsilon
eps = -0.5
w = 2
Y, X = np.mgrid[-w:w:100j, -w:w:100j]
U = Y
V = X - X**3 - eps*Y
speed = np.sqrt(U**2 + V**2)

strm = axs[2].streamplot(X, Y, U, V, density=[0.5, 1], color=U, cmap='autumn')
fig.colorbar(strm.lines)
axs[2].set_title(f'Epsilon of {eps}')

axs[3].set_visible(False)

plt.tight_layout()
plt.savefig('./figures/problem_1_a.png')
plt.show()
```

## Part B

```python
fig, axs = plt.subplots(2, 2, figsize=(8, 6),)
axs = axs.flat

## Positive epsilon
eps = 0.5
w = 5
Y, X = np.mgrid[-w:w:100j, -w:w:100j]
U = Y
V = - np.sin(X) - eps*Y**3
speed = np.sqrt(U**2 + V**2)

strm = axs[0].streamplot(X, Y, U, V, density=[0.5, 1], color=U, cmap='autumn')
axs[0].set_title(f'Epsilon of {eps}')

## Zero epsilon
eps = 0.0
w = 2
Y, X = np.mgrid[-w:w:100j, -w:w:100j]
U = Y
V = - np.sin(X) - eps*Y**3
speed = np.sqrt(U**2 + V**2)

strm = axs[1].streamplot(X, Y, U, V, density=[0.5, 1], color=U, cmap='autumn')
fig.colorbar(strm.lines)
axs[1].set_title(f'Epsilon of {eps}')

## Negative epsilon
eps = -0.5
w = 2
Y, X = np.mgrid[-w:w:100j, -w:w:100j]
U = Y
V = - np.sin(X) - eps*Y**3
speed = np.sqrt(U**2 + V**2)

strm = axs[2].streamplot(X, Y, U, V, density=[0.5, 1], color=U, cmap='autumn')
fig.colorbar(strm.lines)
axs[2].set_title(f'Epsilon of {eps}')

axs[3].set_visible(False)

plt.tight_layout()
plt.savefig('./figures/problem_1_b.png')
plt.show()
```
