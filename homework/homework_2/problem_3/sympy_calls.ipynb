#### Solving the unstable manifold

```python
x, c, d, e = symbols('x a_2 a_3 a_4')
l = symbols('lambda', positive=True)
```

```python
left_hand_side = -l*(c*x**2 + d*x**3 + e*x**4 - x**2)
left_hand_side
```

```python
ce_lhs = collect(expand(left_hand_side), x)
ce_lhs
```

```python
right_hand_side = (2*c*x+3*d*x**2+4*e*x**3) * x * (1-x**2)
right_hand_side
```

```python
ce_rhs = collect(expand(right_hand_side), x)
ce_rhs
```

```python
ce_lhs-ce_rhs
```

```python
collect(ce_lhs-ce_rhs, x)
```

```python
print(latex(collect(ce_lhs-ce_rhs, x)))
```

#### Solving the stable manifold

```python
x, b, c, d, e = symbols('u a_1 a_2 a_3 a_4')
l = symbols('lambda', positive=True)
```

```python
left_hand_side = -l*(1+b*x+c*x**2 + d*x**3 + e*x**4 - (x-1)**2)
left_hand_side
```

```python
ce_lhs = collect(expand(left_hand_side), x)
ce_lhs
```

```python
right_hand_side = (b+2*c*x+3*d*x**2+4*e*x**3) * (x-1) * (1-(x-1)**2)
right_hand_side
```

```python
ce_rhs = collect(expand(right_hand_side), x)
ce_rhs
```

```python
ce_lhs-ce_rhs
```

```python
collect(ce_lhs-ce_rhs, x)
```

```python
print(latex(collect(ce_lhs-ce_rhs, x)))
```
